The Apriori method identifies frequent itemsets in transaction history by selecting those that exceed a minimum support threshold. These frequent itemsets are used to generate association rules, with confidence computed using hybrid dimension association rules. The process reveals correlations among itemsets, allowing for insights into items frequently purchased together by customers within a specific time frame. 
Apriori algorithm doesn’t work on categorical data.

### Identify frequent itemsets, i.e., items or item sets popular in the data sets.

#### Minimum support required in the data set

The support value (s) is the ratio of receipts A ∪ B between sales slips in database D, which corresponds to the ratio of receipts containing both A and B to all receipts. Thus, the support value indicates how often the correlation between the respective products is. 
Support (A->B): P(AUB)

Confidence (c), another value, indicates how much of the receipts containing A also include B. This value is a conditional probability value. The confidence value indicates the strength of the rule.
Confidence (A->B): P(B|A) = P(A∩B)/P(A)

The lift value is used to evaluate the interestingness of a rule more accurately. Specifically, a lift value greater than 1 indicates that the rule has increased interest, while a lift value of 1 suggests that there is no significant association between the items.
Lift (A->B): (PA∩B)/(P(A) . P(B))

Let's consider the itemset {Milk} and {Bread}.
Support Calculations
Support for {Milk, Bread}:
Transactions containing both Milk and Bread: 3 (Transactions 1, 4, and 6)
Total transactions: 9
Support = 3/9 ≈ 0.33
Support for {Milk}:
Transactions containing Milk: 5 (Transactions 1, 3, 4, 6, and 8)
Support = 5/9 ≈ 0.56
Support for {Bread}:
Transactions containing Bread: 6 (Transactions 1, 2, 4, 5, 6, 7, and 9)
Support = 6/9 ≈ 0.67
Confidence Calculation
Confidence of {Milk} → {Bread}:
Confidence = Support(Milk, Bread) / Support(Milk) = 0.33 / 0.56 ≈ 0.6
Lift Calculation
Lift of {Milk} and {Bread}:
Lift = Support(Milk, Bread) / (Support(Milk) × Support(Bread)) = 0.33 / (0.56 × 0.67) ≈ 0.9
Summary
Support:
{Milk, Bread}: 0.33
{Milk}: 0.56
{Bread}: 0.67

Confidence:
{Milk} → {Bread}: 0.6

Lift:
{Milk} and {Bread}: 0.9

```python
# If a very high value of support is set, there won't be any combinations.

minimum_required_support=0.1
```

```python
frequent_itemsets = apriori(selected_columns_df,min_support=minimum_required_support,use_colnames=True)
```

```python
# The following items have support values greater than the minimum support models.

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
## Set rules from the frequent itemsets that match the desired lift.

## Set the required lift

minimum_required_lift = 1.1
rules = association_rules(frequent_itemsets,metric="lift", min_threshold=minimum_required_lift)
```

```python
rules
```

```python
print("Total nuber of rules generated by the algorithm with given support and lift:",len(rules))
```

```python
# Set the minimum threshold for confidence
minimum_required_confidence = 0.5
```

```python
# Filter rules based on confidence and lift threshold
filtered_rules = rules[(rules['confidence'] >= minimum_required_confidence)]
```

```python
filtered_rules
```

```python
print("Total nuber of filtered rules to meet the lift, confidence and support criteria:",len(filtered_rules))
```

```python
# Sorting the rules based on lift, confidence, and support (all in descending order)

sorted_rules = filtered_rules.sort_values(by=['lift','confidence','support'],ascending=False)
```

```python
sorted_rules
```

```python
# Selecting the top 5 rules

top_5_rules = sorted_rules.head(5)
```

```python
# Print only the support, confidence, and lift column

top_5_rules[['antecedents','consequents', 'support', 'confidence', 'lift']]
```
